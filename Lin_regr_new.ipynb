```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
df = pd.read_csv('happiness.csv',sep=';')
df = df.drop(['Overall_rank'], axis='columns', inplace=False)
df
```

```python
df["happiness"] =  np.where(df['Score'] <= 5.8, 0, 1)
```

```python
df = df.drop(['Score'], axis='columns', inplace=False)
```

```python
a = ['GDP_per_capita','Social_support','Fr_to_make_choices','Generosity','Perceptions_of_corruption','happiness']
```

```python
# Splitting the dataset into the Training set and Test set
X = df[a]
y = df['Healthy_life_expectancy']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler().fit(X_train)
X_train = sc_X.transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler() 
scaled_values = scaler.fit_transform(df) 
df.loc[:,:] = scaled_values
```

```python
# Baseline model
import statsmodels.api as sm
lr = sm.Logit(y_train, X_train).fit()
print(lr.summary2())
```
